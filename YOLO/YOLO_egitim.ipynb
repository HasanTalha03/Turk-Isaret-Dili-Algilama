```
import yaml

data = {
    'train': '/content/dataset/train/images',
    'val': '/content/dataset/val/images',
    'nc': 29,
    'names': [
        "i", "s", "o", "c", "u", "A", "B", "C", "D", "E",
        "F", "G", "g", "H", "I", "J", "K", "L", "M", "N",
        "O", "P", "R", "S", "T", "U", "V", "Y", "Z"
    ]
}

with open('/content/dataset/data.yaml', 'w') as f:
    yaml.dump(data, f)
```

```
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

# Eğitimi başlatıyoruz
model.train(
    data='/content/dataset/data.yaml',
    epochs=15,
    imgsz=416,  # Hızlı olması açısından çözünürlüğü biraz düşük tuttuk
    batch=16,
    project='/content/drive/MyDrive/TID_Proje_Yedek', # Herhangi bir bağlantı kopması durumuna karşı verileri drive a yedekliyoruz
    name='Deneme_Final'
)
```
